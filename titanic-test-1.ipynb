```python
titanic_file_path = '/kaggle/input/titanic/train.csv'

titanic_data = pd.read_csv(titanic_file_path)

titanic_data.head()
```

```python
titanic_data.describe()
```

```python
columns = list(titanic_data.columns.values)
print(columns)
```

```python
y = titanic_data.Survived
```

```python
numerical_subset = titanic_data[['Pclass', 'Parch']]

categorical_subset = titanic_data[['Sex', 'Embarked']]
categorical_subset = pd.get_dummies(categorical_subset)

features = pd.concat([numerical_subset, categorical_subset], axis=1)

X = features

print(X.describe)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state = 13)

titanic_model = DecisionTreeRegressor(random_state=13)
titanic_model.fit(X_train, y_train)

train_predictions = titanic_model.predict(X_val)
mae_model_1 = mean_absolute_error(y_val, train_predictions)

print('Model 1 (Pclass, Parch, Sex, Embarked): ',mae_model_1)
```

```python
numerical_subset = titanic_data.Pclass

categorical_subset = titanic_data[['Sex', 'Embarked']]
categorical_subset = pd.get_dummies(categorical_subset)

features = pd.concat([numerical_subset, categorical_subset], axis=1)

X = features

print('Model 2 (Pclass, Sex, Embarked): ', X.describe)
```

```python
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')

numerical_subset = test_data[['Pclass', 'Parch']]

categorical_subset = test_data[['Sex', 'Embarked']]
categorical_subset = pd.get_dummies(categorical_subset)

features = pd.concat([numerical_subset, categorical_subset], axis=1)

X_test = features

test_predictions = titanic_model.predict(X_test).astype(int)
```

```python
output = pd.DataFrame({'PassengerId' : test_data.PassengerId,
                   'Survived' : test_predictions})

output.to_csv('submission.csv',index=False)
```
